```python
# Creating a map dict for oredr status
order_status_dict = {'delivered':8, 'shipped':7, 'canceled':6, 'unavailable':5, 'invoiced':4,
       'processing':3, 'created':2, 'approved':1}
```

```python
# mapping the status variable
df.order_status = df.order_status.map(order_status_dict)
df.head(2)
```

```python
df[['order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date']] = df[['order_purchase_timestamp', 'order_approved_at',
       'order_delivered_carrier_date', 'order_delivered_customer_date',
       'order_estimated_delivery_date']].apply(pd.to_datetime)
```

```python
#Creating new features with timestamps
df["order_purchase_day"] = df["order_purchase_timestamp"].dt.day
df["order_purchase_month"] = df["order_purchase_timestamp"].dt.month
df["order_approved_day"] = df["order_approved_at"].dt.day
df["order_approved_month"] = df["order_approved_at"].dt.month

```

```python
df["order_delivered_carrier_day"] = df["order_delivered_carrier_date"].dt.day
df["order_delivered_customer_day"] = df["order_delivered_customer_date"].dt.day
```

```python
df.drop(["order_purchase_timestamp", "order_approved_at","order_estimated_delivery_date", "review_answer_timestamp" ], axis=1, inplace=True)
df.drop(["order_delivered_carrier_date", "order_delivered_customer_date"], axis=1, inplace= True)
```

```python
seller_state_map = round(df["seller_state"].value_counts()/len(df["seller_state"]),4).to_dict()
customer_state_map = round(df["customer_state"].value_counts()/len(df["customer_state"]),4).to_dict()
```

```python
# Mapping columns
df["seller_state"] = df["seller_state"].map(seller_state_map)
df["customer_state"] = df["customer_state"].map(customer_state_map)
```

```python
df.drop(["product_id", "seller_id"], axis=1, inplace=True)
```

```python
ohe = pd.get_dummies(df["payment_type"])
df = pd.concat([df, ohe], axis=1)
```
